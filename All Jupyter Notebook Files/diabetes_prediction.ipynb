Splitting the Dataset from Training

```python
X = df.drop(columns = 'Outcome' , axis = 1)
Y = df['Outcome']
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 20)
```

```python
print('Training Data  :' , X_train.shape , Y_train.shape)
print('Testing Data : ' , X_test.shape , Y_test.shape)
```

Training our model with the help of RandomForestClassifier Algorithm.

Why to use RandomForestClassification Algorithm?
1. Ans : Random Forest is a powerful and popular machine learning algorithm that is often used for various classification tasks, including diabetes prediction.Random Forest tends to perform well in a wide range of classification problems, including medical diagnosis tasks like diabetes prediction. It can handle both numerical and categorical features, making it suitable for datasets with a mix of data types. It can capture complex relationships between features and the target variable, leading to better accuracy compared to simpler algorithms.

```python
classifier = RandomForestClassifier(n_estimators = 20)
```

1. Here , n_estimators refers to the number of decision trees to be used in the ensemble. Each decision tree in the Random Forest is trained on a different subset of the data, and the final prediction is made by aggregating the predictions of all the trees.
2. Here we have taken n_estimators = 20 it means that we are using 20 decision trees in your Random Forest model. Increasing the number of estimators can lead to better generalization and more robust predictions, up to a point. However, it also increases the computational cost and memory requirements.

```python
classifier.fit(X_train , Y_train)
```

**Now Checking the Accuracy of our Trained Model**

```python
print(accuracy_score(Y_test , classifier.predict(X_test))*100)
```
